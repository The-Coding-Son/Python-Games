```python
import random
suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 
            'Nine':9, 'Ten':10, 'Jack':10, 'Queen':10, 'King':10, 'Ace':1}
```

```python
class Card:
    
    def __init__(self,suit,rank):
        self.suit=suit
        self.rank=rank
        self.value=values[self.rank]
        
    def __str__(self):
        return f"{self.rank} of {self.suit}"
    
```

```python
class Deck:
    
    def __init__(self):
        
        self.all_cards=[]
        for suit in range(4):
            for rank in range(13):
                self.all_cards.append(Card(suits[suit],ranks[rank]))
    def shuffle(self):
        random.shuffle(self.all_cards)
        
    def Deal_One(self):
        return self.all_cards.pop(0)  
```

```python
class Chip:
    
    def __init__(self,count):
        self.count=count
        
    def __valid_bet__(self,request):
        return request<=self.count
    
    def __place_bet__(self,value):
        self.count-=value
        
    def __current_count__(self):
        return f"Player has {self.count} chips left"
```

```python
class Hand:
    
    def __init__(self):
        self.hand=[]
    
    def __add_card__(self,card):
        self.hand.append(card)
        
    def __closeness_factor__(self):
        su=0
        for s in range(len(self.hand)):
            su+=self.hand[s].value    
        return 21-su
        
    def __display__(self):
        for y in range(len(self.hand)):
            print(self.hand[y])
        
        
    def __special_display__(self):
        print("Dealer Hand")
        print(self.hand[0])
        print("Hidden Card")
        print(" ")
```
